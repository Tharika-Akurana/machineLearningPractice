```python
data = pd.read_csv("iris.csv")
data.head()
```

```python
## Species will be the target variable (y) and other 4 raws will be the features(x)

# Store features as x
x = data.iloc[:,1:5]  # iloc - index location
x.head()
```

```python
# Store target variable as y

y = data.iloc[:,-1]
y.head()
```

```python
# feature scalinf the features using standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x = scaler.fit_transform(x)
x[0:5] # display the first 5 values
```

```python
# Divide data into train set and test set

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
## Train the model using KNN algorithm

from sklearn.neighbors import KNeighborsClassifier # Import KNeighborsClaffier Class
model = KNeighborsClassifier(n_neighbors=1) # creating an object for KNeighborsClaffier Class and giving k=1
model.fit(x_train,y_train)
```

```python
# Predict values according to the test set

pred = model.predict(x_test)
pred[0:5] # display first 5 predicted values
```

```python
y_test[0:5] # display first 5 actual test values
```

```python
# Accuracy of the model
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,pred)
accuracy
```

```python
## Find Accuracy using confusin matrix

# Diagonal values give the actual values (true values it's mean real value=predicted value)
# only one value is different (it's mean real value!=predicted value)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, pred)
cm
```

```python
# more details according to the accuracy

result = pd.DataFrame(data=[y_test.values,pred], index=['y_text','pred'])
result.transpose()
```

```python
## Finding a Suitable K value

# Test the model using several K values
correct_sum = []  #creating an array to store values
for i in range(1,20):   # K values from 1 to 19
    model = KNeighborsClassifier(n_neighbors=i) # create an object according to KNeighborsClassifier constructor
    model.fit(x_train,y_train) #train the model
    pred = model.predict(x_test)
    correct = np.sum(pred == y_test)
    correct_sum.append(int(correct))  # if the predicted value = real value, then it add to the list
```

```python
correct_sum
```

```python
result = pd.DataFrame(data=correct_sum)
result.index = result.index+1
result.T
```

```python
# consider K=1
model = KNeighborsClassifier(n_neighbors=8) #(k=8)
model.fit(x_train,y_train)
pred = model.predict(x_test)
```

```python
# find the accuracy
accuracy_score(y_test,pred)
```

```python
# accuracy from confussion matrix
confusion_matrix(y_test,pred)
```
