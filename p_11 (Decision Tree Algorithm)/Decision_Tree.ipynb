```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Kyphosis.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.info() #to find if there is any non-values
```

```python
x = data.drop('Kyphosis',axis=1)  # 'drop' is used for droping a coloumn or raw
```

```python
x.head()  #'x' gives the features
```

```python
y = data['Kyphosis'] #'y' is the target variable
```

```python
y.head() 
```

```python
from sklearn.model_selection import train_test_split # import train_test_split from the sklearn library
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3) # Divide the data set into train set and test set
```

```python
x_train.shape
```

```python
x_test.shape
```

                                          
### DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier # import DecisionTreeClassifier class
model = DecisionTreeClassifier() # Creating a object called 'model'
model.fit(x_train, y_train) # train the model through the object
```

```python
# predict the values using the model
pred = model.predict(x_test)
pred
```

```python
# Real values belong to above predicted values
y_test
```

```python
# checking the accuracy (performance) using accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
# checking the accuracy (performance) using confussion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,pred)
```
