```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
age = np.array([16,24,31,29,40,33,18,36,15,37,19,21,29,31,23,20,34,22,32])
job = np.array([0,0,1,0,1,0,0,1,0,1,0,0,1,1,1,0,1,0,1])
```

```python
plt.scatter(age,job)
plt.xlabel('Age')
plt.show
```

```python
m,c  = np.polyfit(age,job,1)
plt.xlabel('Age')
plt.plot(age,job,'o')
plt.plot(age, m*age + c)
```
